```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import pickle
```

```python
# Load the dataset
df = pd.read_csv('Beetroot-data.csv')
df.head()
```

```python
# EDA: Check for null values and duplicates
print("Null values in each column:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Handle missing values and duplicates
df = df.dropna().drop_duplicates()
```

```python
# Basic statistics
print("Descriptive Statistics:\n", df.describe())
```

```python
# Visualize the distribution of Modal Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Modal Price (Rs./Quintal)'], bins=30, kde=True)
plt.title('Distribution of Modal Prices')
plt.xlabel('Modal Price (Rs./Quintal)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Modal Prices by District
plt.figure(figsize=(12, 6))
sns.boxplot(x='District Name', y='Modal Price (Rs./Quintal)', data=df)
plt.title('Boxplot of Modal Prices by District')
plt.xticks(rotation=45)
plt.xlabel('District Name')
plt.ylabel('Modal Price (Rs./Quintal)')
plt.show()
```

```python
# Boxplot for Modal Prices by Market
plt.figure(figsize=(12, 6))
sns.boxplot(x='Market Name', y='Modal Price (Rs./Quintal)', data=df)
plt.title('Boxplot of Modal Prices by Market')
plt.xticks(rotation=45)
plt.xlabel('Market Name')
plt.ylabel('Modal Price (Rs./Quintal)')
plt.show()
```

```python
# Correlation heatmap
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])  # Select numeric types
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Encode categorical features
district_encoder = LabelEncoder()
market_encoder = LabelEncoder()
```

```python
df['District Name'] = district_encoder.fit_transform(df['District Name'])
df['Market Name'] = market_encoder.fit_transform(df['Market Name'])
```

```python
# Convert 'Price Date' to datetime
df['Price Date'] = pd.to_datetime(df['Price Date'], format='%d-%b-%y', errors='coerce')
```

```python
# Extract Year and Month from 'Price Date'
df['Year'] = df['Price Date'].dt.year
df['Month'] = df['Price Date'].dt.month
```

```python
# Select features and target variable
X = df[['District Name', 'Market Name', 'Year', 'Month']]
y = df['Modal Price (Rs./Quintal)']
```

```python
# Scale the target variable
scaler = StandardScaler()
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1)).flatten()
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.2, random_state=42)
```

```python
# Model training
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Model prediction
y_pred = rf.predict(X_test)
```

```python
# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

```
